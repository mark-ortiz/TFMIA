## Uniendo la data de ambos dataframes

Estoy explorando la idea de "UNIR" los datos relevante de ambos dataframes para quedarme con un solo dataframe, el cual contenga toda la data relevante, aunque si que estaran "repetidos" muchos datos, ya que le correponden a diversas lineas del evento, la misma data de aafit, bbfit y del muón.

Aquí la clave es la metadata, y lo pienso hacer por runid, frameid e interactionid

Definí dos funciones que me permitirán trabajar con runid,frameid e interactionid. Se muestran a continuación.

```python
### FUNCIONES DE APOYO

#Función para contar el # de eventos con respecto a runid e interactionid, es decir en un "archivo csv"
def func_calc_eventos(runid,interactionid):
    otrodf = df_aafit[df_aafit['runID'] == runid]
    otrodf = otrodf[otrodf['interactionID']==interactionid]
    return(len(otrodf))

#Función que sirve para buscar un runid,frameid e interactionid en específico
#con base a lo que queramos hacer match del df aafit
def func_busca_match_hits(runid,frameid,interactionid):
    tempdf = df_hits[df_hits['runID']==runid]
    tempdf = tempdf[tempdf ['frameID']==frameid]
    indices = tempdf.index.to_numpy()
    lineas=len(tempdf)
    return(lineas,indices)
```

```python
#Si nos interesa conocer cuantos eventos tiene cierto archivo CSV (runID con interactionID)
#podemos hacerlo de la siguiente manera.

#Por ejemplo para el primer evento:
func_calc_eventos(df_aafit.iloc[0][0],2)
```

```python
#Con esto agrego las columnas que "faltan" en el dataframe de los hits.
a=['aafit_azimut','aafit_zenit','aafit_lambda','aafit_beta','bbfit_azimut',
   'bbfit_zenit','bbfit_quality','muon_azimut','muon_zenit','muon_energy']
for i in a:
    df_hits[i]=np.nan
```

```python
##################################################################################
#### CODIGO PARA ASIGNAR LOS VALORES CORRESPONDIENTES DEL DATAFRAME AAFIT    #####
#### Y "PEGARLOS" AL DATAFRAME DE LOS HITS, ES COMO COPIAR Y PEGAR COLUMNAS  #####
##################################################################################

#Como "llenar" los valores vacios en df_hits con los valores que le corresponden
#del df_aafit, en este caso son 19 líneas porque es el primer evento

#Debo hacer bucles anidados, el primero para recorrer cada línea de aafit
#y otro para asignar los valores de cada evento con base en runid,frameid e interactionid

for j in range(len(df_aafit)):
    rid,fid,iid = df_aafit.iloc[j][0], df_aafit.iloc[j][1], df_aafit.iloc[j][3]
    #Calculo el numero de lineas y los indices de df_hits
    lineas, indices = func_busca_match_hits(rid,fid,iid)
    #print(lineas,indices)
    for i in indices: 
        df_hits.iat[i,17] = df_aafit.iloc[j][4]
        df_hits.iat[i,18] = df_aafit.iloc[j][5]
        df_hits.iat[i,19] = df_aafit.iloc[j][6]
        df_hits.iat[i,20] = df_aafit.iloc[j][7]
        df_hits.iat[i,21] = df_aafit.iloc[j][8]
        df_hits.iat[i,22] = df_aafit.iloc[j][9]
        df_hits.iat[i,23] = df_aafit.iloc[j][10]
        df_hits.iat[i,24] = df_aafit.iloc[j][11]
        df_hits.iat[i,25] = df_aafit.iloc[j][12]
        df_hits.iat[i,26] = df_aafit.iloc[j][13]
        
#### NOTA: LA EJECUCIÓN DE ESTE CÓDIGO, TOMÓ UNA SEMANA ENTERA!!!!!!!
```

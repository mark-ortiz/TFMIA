```python
import pandas as pd
import math
import timeit

#path = "U:\MASTER INT. ARTIFICIAL\TFM_DATA\CLEANDATA\\" #Path para mi laptop en casa
path = "F:\\DATA_TFM\\CLEANDATA\\" #Path para mi PC en el trabajo

#archivo = "data_aafit_bbfit_muon.csv"
archivo = "data_best_hits_full.csv"

file = path+archivo
```

```python
#Necesito calcular azimut y zenit para el dataframe que ya obtuve de los hits

df = pd.read_csv(file)

#En cargar el df en la laptop tarda alrededor de 5 min, se debe tener cuidado, verificando el CPU de la máquina,
#llega a calentarse hasta 70°C, lo que en una laptop puede dañarla si se mantiene mucho tiempo así.
#en la PC Desktop del trabajo solo llega hasta 35°C lo cual es muy bueno.
```

```python
#Generando un "chunk" del df principal para solo hacer la demostración de insertar azimut y zenit

#### Hay que ejecutar la celda de validación más abajo **** mucho ojo con esto para la prueba ****
```

```python
#Calculo del azimut y zenit para cada row del dataframe

col_azimut=[];col_zenit=[]

for i in range(len(df)):
    if i % 1000000 == 0:
        print (i)
    #Obteniendo los valores
    X = df.iloc[i]['X']; Y = df.iloc[i]['Y']; Z = df.iloc[i]['Z']
    #print("%.6f" %X, "%.6f" %Y, "%.6f" %Z)
    #print("X:%.4f", %X)
    #Calculando el azimut de la fila
    azimut=math.atan2(Y,X)
    #Calculando el zenit de la fila
    zenit = math.acos(Z)
    #print("azimut: %.6f" %azimut, "zenit: %.6f" %zenit)
    col_azimut.append(azimut)
    col_zenit.append(zenit)

print(len(col_azimut),len(col_zenit))

#No es bueno imprimir la salida de i en cada bucle mejor cada millón, ya que
#se alenta mucho el notebook, pienso que es porque se acaba la memoria de video al etar imprimiendo i
```

```python
#Insertando la data de azimut y zenit al df completo de 60M de líneas
df.insert(15,'azimut',col_azimut)
df.insert(16,'zenit',col_zenit)
```

```python
#Grabamos el df ya con la data de azimut y zenit como un CSV

#outputpath = "U:\MASTER INT. ARTIFICIAL\TFM_DATA\CLEANDATA\\"
outputpath = "F:\DATA_TFM\CLEANDATA\\"
    
df.to_csv(outputpath+"data_best_hits_completo.csv", index = None)#, header=None)
```
